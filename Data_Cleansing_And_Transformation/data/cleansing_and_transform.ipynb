Writing Transformation Functions

```python
def get_transformed_data(df_traffic_crashes, df_traffic_crash_vehicle):
    under_threshold_removed = df_traffic_crashes.dropna(axis='index', thresh=2, inplace=False)
    under_threshold_rows = df_traffic_crashes[~df_traffic_crashes.index.isin(under_threshold_removed.index)]
    df_traffic_crashes.fillna(value={'report_type':'ON SCENE'}, inplace=True)

    # merge crashes and vehicles
    df = df_traffic_crashes.merge(df_traffic_crash_vehicle, how='left', on='crash_record_id', suffixes=('_left','_right'))
    df_agg = df.groupby('vehicle_type').agg({'crash_record_id':'count'}).reset_index()

    # transform column names for output data
    vehicle_mapping = {'vehicle_type':'vehicletypes'}
    df_agg = df_agg.rename(columns=vehicle_mapping)
    return df_agg
```

```python
# read data from data source
def read_datasources(source_name):
    df = pd.read_csv(source_name)
    return df
```

```python
# Drop rows with null values
def drop_row_with_null_values(df):
    under_threhold_removed = df.dropna(axis='index', thresh=2, inplace=False)
    df = df[~df.index.isin(under_threhold_removed.index)]
    return df
```

```python
# Fill missing values
def fill_missing_values(df):
    df = df.fillna(value={'report_type':'ON SCENE'})
    return df
```

```python
# merge Dataframes
def merge_data_frames(df_vehicles, df_crashes):
    df = df_crashes.merge(df_vehicles, how='left', on='crash_record_id', suffixes=('_left', '_right'))
    return df
```

```python
# Rename columns
def rename_columns(df):
    vehicle_mapping = {'vehicle_type' : 'vehicletypes'}
    df = df.rename(columns=vehicle_mapping)
    return df
```
